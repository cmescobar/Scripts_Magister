# Zero padding effect

```python
%matplotlib notebook
signal = np.array([np.cos(i * 2 * np.pi / 50) for i in range(100)])

plt.subplot(2,2,1)
plt.plot(signal)

plt.subplot(2,2,2)
plt.plot(abs(np.fft.fft(signal)))

signal_app = np.append(signal, [0] * 50)

plt.subplot(2,2,3)
plt.plot(signal_app)

plt.subplot(2,2,4)
plt.plot(abs(np.fft.fft(signal_app)))

plt.show()
```

# IDFT Zero padding frequency signal

```python
%matplotlib notebook

plt.plot(np.fft.ifft((np.fft.fft(signal_app))))
plt.show()
```
