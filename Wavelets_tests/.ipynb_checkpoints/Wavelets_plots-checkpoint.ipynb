```python
# Recorriendo la lista de los wavelets discretos 
for wav_type in pywt.wavelist(kind='discrete'):
    # Definición de la familia
    try:
        num = re.search('\d', wav_type).start()
        family = wav_type[:num]
    except:
        family = wav_type

    # Definición de la carpeta dónde se guardará la información
    folder_data = f'Wavelets_plots/Family_{family}'

    # Preguntar si es que la carpeta que almacenará los sonidos se ha
    # creado. En caso de que no exista, se crea una carpeta
    if not os.path.isdir(folder_data):
        os.makedirs(folder_data)

    print(f"Plotting Wavelet {wav_type}...")
    wavelet_info = pywt.Wavelet(wav_type).wavefun(level=5)

    plt.figure(figsize=(9,5))
    plt.subplot(1,2,1)
    plt.plot(wavelet_info[-1], wavelet_info[0])
    plt.title('Función de escalamiento')

    plt.subplot(1,2,2)
    plt.plot(wavelet_info[-1], wavelet_info[1])
    plt.title('Función Wavelet')

    # Título
    plt.suptitle(f'Wavelets {wav_type} ({len(wavelet_info[-1])} Muestras)')

    # Guardando
    plt.savefig(f'{folder_data}/{wav_type}.png')

    #Cerrando
    plt.close()

    print('Completed!\n')
```
